```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
import time

# Replace these with your Kijiji credentials and ad details
username = 'Butter'
password = 'Jr9058454326!'
ad_title = 'Brand new basement apartment for rent'
ad_description = 'New basement apartment for rent. Stainless steel appliances coming soon.'
ad_price = '2000'
location = 'Brampton'

# List of paths to pictures you want to upload
picture_paths = ['C:\Users\jessi\Downloads\picture1.jpg',
                 '/path/to/picture2.jpg',
                ]


# Path to your webdriver (e.g., chromedriver.exe)
webdriver_path = '/path/to/chromedriver'

# URL of the Kijiji website
kijiji_url = 'https://www.kijiji.ca/'

# Start the web browser
driver = webdriver.Chrome(executable_path=webdriver_path)

try:
    # Open Kijiji website
    driver.get(kijiji_url)

    # Find and click on the "Post Ad" button or link
    post_ad_button = driver.find_element_by_link_text('Post Ad')
    post_ad_button.click()

    # Log in (assuming you already have an account)
    login_button = driver.find_element_by_link_text('Sign In')
    login_button.click()

    username_field = driver.find_element_by_name('emailOrNickname')
    password_field = driver.find_element_by_name('password')

    username_field.send_keys(username)
    password_field.send_keys(password)

    password_field.send_keys(Keys.RETURN)

    # Fill in ad details
    ad_title_input = driver.find_element_by_id('postad-title')
    ad_description_input = driver.find_element_by_id('pstad-descrptn')
    ad_price_input = driver.find_element_by_id('priceAmount')
    location_input = driver.find_element_by_id('pstad-map-location-search')

    ad_title_input.send_keys(ad_title)
    ad_description_input.send_keys(ad_description)
    ad_price_input.send_keys(ad_price)
    location_input.send_keys(location)

    # Upload pictures
    for i, picture_path in enumerate(picture_paths):
        picture_input = driver.find_element_by_id(f'pstad-image-uploader-input-{i+1}')
        picture_input.send_keys(picture_path)

    # Submit the ad
    submit_button = driver.find_element_by_id('pstad-btn-create')
    submit_button.click()

    # Wait for some time to let the ad submission complete (you may need to adjust this)
    time.sleep(5)

finally:
    # Close the browser window
    driver.quit()
```
